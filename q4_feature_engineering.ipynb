## Your Approach

1. **Create derived features** - Differences, ratios, interactions between variables (watch for division by zero)
2. **Calculate rolling windows** - Use `.rolling()` on predictor variables to capture temporal patterns

   ⚠️ **Data Leakage Warning:** Do not create ANY features that use your target variable - this includes rolling windows, differences, ratios, or interactions involving the target. For example, if predicting Air Temperature, do not create `air_temp * humidity` or `air_temp - wet_bulb`. Only derive features from other predictor variables.

3. **Create categorical features** - Bin continuous variables if useful (optional)
4. **Check for infinity values** - Ratios can produce infinity; replace with NaN and handle appropriately
5. **Document and save** - Remember to `reset_index()` before saving CSVs

---

## Decision Points

- **Derived features:** What relationships might be useful? Temperature differences? Ratios? Interactions between variables?
- **Rolling windows:** What window size makes sense? 7 hours? 24 hours? Consider the temporal scale of your data. For hourly data, 7-24 hours captures daily patterns.
- **Time-based aggregations:** Aggregate by hour? Day? Week? What temporal granularity is useful for your analysis?

````python
df = pd.read_csv(
    "output/q3_wrangled_data.csv",
    parse_dates=["Measurement Timestamp"],
    index_col="Measurement Timestamp"
)

df_fe = df.copy()

# =========================
# Safe Derived Features (Predictor-only)
# =========================

# Wind energy proxy
df_fe["wind_speed_squared"] = df_fe["Wind Speed"] ** 2

# Rain intensity indicator
df_fe["is_raining"] = (df_fe["Interval Rain"] > 0).astype(int)

# Pressure change proxy (first difference)
df_fe["pressure_change"] = df_fe["Barometric Pressure"].diff()

# Solar efficiency proxy (normalized solar radiation)
df_fe["solar_radiation_normalized"] = df_fe["Solar Radiation"] / df_fe["Solar Radiation"].max()

# Wind chill interaction proxy (safe — no air temp used)
df_fe["wind_humidity_interaction"] = df_fe["Wind Speed"] * df_fe["Humidity"]

# Replace any infinities created by division
df_fe = df_fe.replace([np.inf, -np.inf], np.nan)

print("✅ Derived features created")
````

````python
#Creating rolling variables for predictors - Wind speed,humidity and pressure - 7 hrs with 24hr window for wind speed and humidity
rolling_df = df_fe.copy()

# 7-hour rolling means (short-term dynamics)
rolling_df["wind_speed_rolling_7h"] = rolling_df["Wind Speed"].rolling(window=7).mean()
rolling_df["humidity_rolling_7h"] = rolling_df["Humidity"].rolling(window=7).mean()
rolling_df["pressure_rolling_7h"] = rolling_df["Barometric Pressure"].rolling(window=7).mean()

# 24-hour rolling means (daily patterns)
rolling_df["wind_speed_rolling_24h"] = rolling_df["Wind Speed"].rolling(window=24).mean()
rolling_df["humidity_rolling_24h"] = rolling_df["Humidity"].rolling(window=24).mean()

print("✅ Rolling window features created")
````

````python
#Resetting Index and Savinf features into csv
df_fe.reset_index().to_csv("output/q4_features.csv", index=False)
print("✅ output/q4_features.csv saved")
````

````python
q4_rolling_features = rolling_df.reset_index()[[
    "Measurement Timestamp",
    "wind_speed_rolling_7h",
    "humidity_rolling_7h",
    "pressure_rolling_7h",
    "wind_speed_rolling_24h",
    "humidity_rolling_24h"
]]

q4_rolling_features.to_csv("output/q4_rolling_features.csv", index=False)

print("✅ output/q4_rolling_features.csv saved")
display(q4_rolling_features.head())
````
